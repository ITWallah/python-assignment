```python
## Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

```python
ArithmeticError have 3 types :
    
--> ZeroDivisionError
--> OverflowError
--> FloatingPointError
```

```python
print("this is example of Zero division error")

try:
    10/0
except ArithmeticError as e:
    print(e)
```

```python
print ("This is example of over flow error")

j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(e)
```

```python
## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
-> The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively
```

```python
print("this is the example of index error")

x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(e)
```

```python
print("this is the example of key error")

countries = {'Italy': 'Rome', 'Poland': 'Warsaw', 'UK': 'London'}
  
try:
    countries["India"]
except KeyError as e:
    print(e)
    #print('The capital of France is unknown')
```
